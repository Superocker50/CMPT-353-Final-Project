```python
def validate_price(p):
    price_str = str(p)
    price_str = price_str.replace("$", "")
    price_str = price_str.replace("Everyone", "0")
    return float(price_str)
```

```python
apps["Price"] = apps.apply(lambda row: validate_price(row["Price"]), axis=1) 
```

```python
# remove rating above 5
apps = apps[apps["Rating"] <= 5]

apps[["Reviews", "Price"]] = apps[["Reviews", "Price"]].apply(pd.to_numeric)
```

```python
apps.info()
```

```python
f = plt.figure(figsize=(15, 11))
plt.matshow(apps.corr(), fignum=f.number)
plt.xticks(range(apps.select_dtypes(['number']).shape[1]), apps.select_dtypes(['number']).columns, fontsize=14, rotation=45)
plt.yticks(range(apps.select_dtypes(['number']).shape[1]), apps.select_dtypes(['number']).columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16);

plt.show()
```
